# Data Dictionary

Columns | Meaning
------ | ---------
account_id | Unique customer/company identifier
monthly_login_freq | Number of logins per month
num_support_ticket | Support requests count
contract_length_months | Subscription duration
avg_product_usage_hours | Product usuage time
num_users | Users under the account
nps_score | Net Promoter Score
has_admin_activity | Whether admins are active (0:No, 1:Yes)
last_activity_days_ago | Recency metric
avg_session_length | Average session duration
feature_adoption_score | Adoption of platform features
customer_satisfaction_score | Satisfaction rating
city | Customer location
gender | Contact-user gender
churned | Target variable (0:Retained, 1:Churned)
